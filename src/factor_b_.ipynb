```python
import pandas as pd
import numpy as np

data = pd.read_csv('/Users/abby/Projects/time_series_analysis-main/data/futures/raw/AMZN.csv')
```

```python
data = data.sort_values(by='timestamp')
data.reset_index(drop=True, inplace=True)
```

```python
def factor_b(zf, price, period):
    zf_mean = zf.rolling(period).mean()
    price_mean = price.rolling(period).mean()
    b = zf_mean*price_mean
    return b
```

```python
zf = data['high']/data['low']
zhenfu = pd.DataFrame(zf)
zhenfu.columns = ['zhenfu']
f = data['high']/data['low']*data['adjusted_close']
factor = pd.DataFrame(f)
factor.columns = ['factor']
```

```python
result = pd.concat([data, zhenfu], axis=1)
result = pd.concat([result, factor],axis=1)
print(result[:50])
```

```python
a = result['zhenfu']
b = result['adjusted_close']
factor_b = factor_b(a,b, 20)
```

```python
fb = pd.DataFrame(factor_b)
fb.columns = ['factor_b']
result = pd.concat([result,fb],axis=1)
```

```python
d = result[20:]
c = d['factor']-d['factor_b']
signal = np.sign(c)
thesig = signal.shift(1)
```

```python
ss = pd.DataFrame(thesig)
ss.columns = ['signal']
```

```python
da = pd.concat([d,ss],axis=1)
print(da)
```

```python
df = da.drop(['Unnamed: 0'], axis=1)
```
